### Step 8. 2개 이상의 Canned Soda 주문은 몇 건 일어났나요?

Hint: 불리언 인덱싱
  - 두 개의 필터를 만든 뒤 & 조건으로 묶으세요
    - item_name이 Canned Soda
    - quantity가 2 이상 (quantity >= 2)
  - 불리언 인덱싱 후 행의 갯수를 세면 됩니다. (len() 또는 .count())



```
len(chipo[(chipo.item_name == 'Canned Soda') & (chipo.quantity >= 2)])
```
